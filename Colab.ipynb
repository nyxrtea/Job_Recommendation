```python
import re
import string

df = pd.read_csv('/content/dataset_cleaning.csv')

# Fungsi untuk membersihkan teks dan mengganti titik koma dengan koma
def clean_text(text):
    # Hapus mention seperti @username
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    # Hapus hashtag seperti #hashtag
    text = re.sub(r'#\w+', '', text)
    # Hapus retweet seperti RT
    text = re.sub(r'RT[\s]+', '', text)
    # Ganti semua tanda baca kecuali koma dengan spasi
    text = text.translate(str.maketrans({p: ' ' for p in string.punctuation if p not in [',', ';']}))
    # Ganti titik koma dengan koma
    text = text.replace(';', ',')
    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Terapkan fungsi clean_text ke kolom 'Interests'
df['Interests'] = df['Interests'].apply(clean_text)
df['Skills'] = df['Skills'].apply(clean_text).str.lower()
df['Interests'] = df['Interests'].str.lower()  # Ubah ke huruf kecil
df['UG Specialization (Major)']=df['UG Specialization (Major)'].str.lower()
df['Certification Course Title']=df['Certification Course Title'].str.lower()
df['First Job Title']=df['First Job Title'].str.lower()
df['Masters Field']=df['Masters Field'].str.lower()
df['Certification Courses']=df['Certification Courses'].str.lower()
df['Working Status']=df['Working Status'].str.lower()
df['Gender']=df['Gender'].str.lower()

df
```

```python
df['Gender'] = df['Gender'].replace({'female': 1, 'male': 0,'prefer not to say':2}) #mengubah value
df['Certification Courses'] = df['Certification Courses'].replace({'yes': 1, 'no': 0})
df['Working Status'] = df['Working Status'].replace({'yes': 1, 'no': 0})

df
```

```python
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# KNN Manual
class KNN:
    def __init__(self, k=17):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def _euclidean_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))

    def _predict(self, x):
        distances = [self._euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_neighbors = [self.y_train[i] for i in k_indices]
        most_common = Counter(k_neighbors).most_common(1)
        return most_common[0][0]

    def predict(self, X):
        return [self._predict(x) for x in np.array(X)]

# 1. Dataset: Pastikan dataset tersedia dalam variabel df
df = pd.read_csv('dataset_fiks.csv')

# 2. Menggabungkan kolom teks menjadi satu kolom untuk analisis TF-IDF
df['combined_text'] = df['Interests'] + ' ' + df['Skills'] + ' ' + df['Certification Course Title'] + ' ' + df['UG Specialization (Major)']

# 3. Menggunakan TF-IDF untuk mengubah teks menjadi fitur
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english', ngram_range=(1, 2))  # Unigrams dan bigrams
tfidf_features = vectorizer.fit_transform(df['combined_text']).toarray()
tfidf_df = pd.DataFrame(tfidf_features, columns=vectorizer.get_feature_names_out())

# 4. Menggabungkan fitur TF-IDF dengan fitur numerik lainnya
df = pd.concat([df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)

# 5. Menghapus kolom teks asli dan kolom sementara
df = df.drop(columns=['Interests', 'Skills', 'Certification Course Title', 'UG Specialization (Major)', 'combined_text'])

# 6. Pisahkan Fitur dan Target
X = df.drop(columns=['Mapped Category'])  # Fitur
y = df['Mapped Category']  # Target

# Ubah kolom kategorikal menjadi numerik
for col in X.columns:
    if X[col].dtype == 'object':  # Jika kolom bertipe string atau object
        X[col] = X[col].astype('category').cat.codes

# Menggunakan MinMaxScaler menghindari dominasi fitur
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Simpan parameter scaler
joblib.dump(scaler, 'normalization_params.joblib')

# 8. Split Data
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y.values, test_size=0.2, random_state=42)

# 9. Menggunakan PCA untuk reduksi dimensi
pca = PCA(n_components=0.95)  # Mempertahankan 95% varians
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# 10. Tuning Parameter K untuk Model KNN
best_accuracy = 0
best_k = None

for k in range(3, 21):  # Mencoba nilai k dari 3 hingga 20
    knn_model = KNN(k=k)
    knn_model.fit(X_train_pca, y_train)
    y_pred = knn_model.predict(X_test_pca)
    accuracy_value = accuracy_score(y_test, y_pred)

    print(f"Accuracy with k={k}: {accuracy_value:.2f}")

    if accuracy_value > best_accuracy:
        best_accuracy = accuracy_value
        best_k = k

# 11. Menampilkan hasil terbaik
print(f"\nBest accuracy found with k={best_k}: {best_accuracy:.2f}")

# 12. Menyimpan Model KNN dengan k terbaik
knn_model_best = KNN(k=best_k)
knn_model_best.fit(X_train_pca, y_train)
joblib.dump(knn_model_best, "best_knn_model.joblib")

# 13. Menghitung dan Menampilkan Akurasi Model Terbaik
y_best_pred = knn_model_best.predict(X_test_pca)
best_model_accuracy = accuracy_score(y_test, y_best_pred)
print(f"Accuracy of the best KNN model with k={best_k}: {best_model_accuracy:.2f}")
```

```python
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# KNN Manual
class KNN:
    def __init__(self, k=17):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def _euclidean_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))

    def _predict(self, x):
        distances = [self._euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_neighbors = [self.y_train[i] for i in k_indices]
        most_common = Counter(k_neighbors).most_common(1)
        return most_common[0][0]

    def predict(self, X):
        return [self._predict(x) for x in np.array(X)]

# Fungsi untuk menyimpan model ke file joblib
def save_model(model, filename):
    with open(filename, 'wb') as file:
        joblib.dump(model, file)

# Fungsi untuk memuat model dari file joblib
def load_model(filename):
    with open(filename, 'rb') as file:
        return joblib.load(file)

# 1. Dataset: Pastikan dataset tersedia dalam variabel df
df = pd.read_csv('dataset_fiks.csv')

# 2. Menggabungkan kolom teks menjadi satu kolom untuk analisis TF-IDF
df['combined_text'] = df['Interests'] + ' ' + df['Skills'] + ' ' + df['Certification Course Title'] + ' ' + df['UG Specialization (Major)']

# 3. Menggunakan TF-IDF untuk mengubah teks menjadi fitur
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english', ngram_range=(1, 2))  # Unigrams dan bigrams
tfidf_features = vectorizer.fit_transform(df['combined_text']).toarray()
tfidf_df = pd.DataFrame(tfidf_features, columns=vectorizer.get_feature_names_out())

# 4. Menggabungkan fitur TF-IDF dengan fitur numerik lainnya
df = pd.concat([df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)

# 5. Menghapus kolom teks asli dan kolom sementara
df = df.drop(columns=['Interests', 'Skills', 'Certification Course Title', 'UG Specialization (Major)', 'combined_text'])

# 6. Pisahkan Fitur dan Target
X = df.drop(columns=['Mapped Category'])  # Fitur
y = df['Mapped Category']  # Target

# Ubah kolom kategorikal menjadi numerik
for col in X.columns:
    if X[col].dtype == 'object':  # Jika kolom bertipe string atau object
        X[col] = X[col].astype('category').cat.codes

# Menggunakan MinMaxScaler
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Simpan parameter scaler
joblib.dump(scaler, 'normalization_params.joblib')

# 8. Split Data
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y.values, test_size=0.2, random_state=42)

# 9. Menggunakan PCA untuk reduksi dimensi
pca = PCA(n_components=0.95)  # Mempertahankan 95% varians
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# 10. Membuat dan Melatih Model KNN dengan k=17 (contoh k)
knn_model = KNN(k=17)
knn_model.fit(X_train_pca, y_train)

# 11. Evaluasi Model
y_pred = knn_model.predict(X_test_pca)

# 12. Menghitung dan Menampilkan Akurasi
accuracy_value = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy_value:.2f}")

# 13. Menyimpan Model ke File
save_model(knn_model, "knn_model.joblib")

# 14. Memuat Model dan Menguji Ulang
loaded_knn_model = load_model("knn_model.joblib")
loaded_y_pred = loaded_knn_model.predict(X_test_pca)
loaded_accuracy_value = accuracy_score(y_test, loaded_y_pred)
print(f"Accuracy of loaded model: {loaded_accuracy_value:.2f}")

# Simpan TF-IDF Vectorizer ke file
joblib.dump(vectorizer, 'vectorizer.joblib')
```
